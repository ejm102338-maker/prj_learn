```python
from sklearn.preprocessing import LabelEncoder

# traing 데이터만 학습 시켜야 한다.
items = ['TV','냉장고','전자레인지','컴퓨터','선풍기','냉장고','TV']

encoder = LabelEncoder()

result = encoder.fit_transform(items)
result
```

```python
encoder.classes_
```

```python
# 결과를 숫자로 반환, 문자로 변환하는 명령
result = encoder.inverse_transform([3,1,2])
result
```
