```python
import pandas as pd
import numpy as np

user_df = pd.read_csv('./data/ratings_nohead.csv')
movie_df = pd.read_csv('./data/movies.csv')
user_df.columns = ["userid","movieId","rating","timestamp"]
user_df.head()
```

```python
movie_df.head()
```

```python
user_df.info()
```

```python
movie_df.info()
```

```python
movie_ratings_df = pd.merge(
    user_df,
    movie_df,
    on='movieId'
)
```

```python
movie_ratings_df.head()
```

```python
user_ratings_df = movie_ratings_df.pivot_table(index='userid',columns='title',values='rating',fill_value=0) 
```

```python
user_ratings_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cos_user_df = cosine_similarity(user_ratings_df,user_ratings_df)
cos_user_df
```

```python
cos_df = pd.DataFrame(cos_user_df,index=user_ratings_df.index,columns=user_ratings_df.index)

cos_df.head()
```

```python
cos_df.iloc[0].sort_values(ascending=False).head(10)
```
