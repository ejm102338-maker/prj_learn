```python
import pandas as pd
import numpy as np

ratings_df = pd.read_csv('data/ratings.csv')
movies_df = pd.read_csv('data/movies.csv')
```

```python
movie_ratings_df = pd.merge(ratings_df,movies_df, on='movieId')
movie_ratings_df
```

```python
user_ratings_df = movie_ratings_df.pivot_table(index='userId',columns='title',values='rating')
user_ratings_df = user_ratings_df.fillna(0)
user_ratings_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity

user_base_result = cosine_similarity(user_ratings_df,user_ratings_df)
user_base_result
```

```python
user_base_df = pd.DataFrame(user_base_result, index=user_ratings_df.index , columns=user_ratings_df.index)
user_base_df
```
